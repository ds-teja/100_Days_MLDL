```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Let's see the available datasets in seaborn
sns.get_dataset_names()
```

```python
# Let's use the restaurant tips data available in seaborn
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## 1. Categorical Plots

### 1.1 Count Plot

Displays the counts of observations in each category. Suitable for visualizing the frequency or distribution of categorical data.

```python
# Count Plot
plt.figure(figsize=(8, 5))
sns.countplot(x="day", data=tips, palette="Set3")
plt.title("Count of Tips by Day of the Week")
plt.show()
```

### 1.2 Swarm Plot

Displays individual data points along a categorical distribution. This is Useful for visualizing the distribution and density of data points within categories.

```python
# Swarm Plot
plt.figure(figsize=(8, 5))
sns.swarmplot(x="day", y="total_bill", data=tips, palette="viridis")
plt.title("Total Bill Distribution by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

### 1.3 Point Plot

Shows point estimates and confidence intervals as vertical lines. This is Useful for visualizing the central tendency and variability of numerical data across categories.

```python
# Point Plot
plt.figure(figsize=(8, 5))
sns.pointplot(x="day", y="total_bill", data=tips, ci="sd", palette="pastel")
plt.title("Average Total Bill by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

This is used to create various types of categorical plots and allows for easy customization of the type of categorical plot (e.g., bar plot, point plot, box plot) using the kind parameter.

### 1.4 Categorical Box Plot
Visualizes the distribution of a quantitative variable for different categories. Displays quartiles (median, upper and lower quartiles), potential outliers, and overall data spread.

```python
# Categorical Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x="time", y="total_bill", data=tips, palette="coolwarm")
plt.title("Total Bill Distribution by Meal Time")
plt.xlabel("Meal Time")
plt.ylabel("Total Bill ($)")
plt.show()
```

## 1.5 Categorical ViolinPlot
Displays individual data points along a categorical distribution. Minimizes overlap of data points, making it useful for visualizing the density of data.

```python
# Categorical Violin Plot
plt.figure(figsize=(10, 8))
sns.violinplot(x="day", y="total_bill", data=tips, palette="Set2")
plt.title("Total Bill Distribution by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

### 1.6 Cat Plot

```python
# Box Plot using cat plot
plt.figure(figsize=(8, 5))
sns.catplot(x="day", y="total_bill", data=tips, kind="box", palette="coolwarm")
plt.title("Total Bill Distribution by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
# Violin Plot using cat plot
plt.figure(figsize=(8, 5))
sns.catplot(x="time", y="total_bill", data=tips, kind="violin", palette="Set2")
plt.title("Total Bill Distribution by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Total Bill ($)")
plt.show()
```
